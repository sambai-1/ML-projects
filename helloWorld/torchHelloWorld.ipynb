```python
import torch
import numpy as np
import torch.nn as nn
import torch.optim as optim
```

```python
xs = torch.tensor([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0]).unsqueeze(1)  # shape: (6, 1)
ys = torch.tensor([-2.0, 1.0, 4.0, 7.0, 10.0, 13.0]).unsqueeze(1)  # shape: (6, 1)

model = nn.Linear(1, 1)
loss_base= nn.MSELoss()
optimizer = optim.SGD(model.parameters(), lr=0.01)
```

```python
for epoch in range(1000):
    # Forward pass
    outputs = model(xs)
    loss = loss_base(outputs, ys)

    # Backward pass and optimization
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    print(epoch, loss.item())
```
